**1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(20)
y = np.random.rand(20) 

plt.scatter(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Scatter Plot")

plt.show()
```

**2. Write a Python program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other.**

```python
x = np.random.rand(20) 
y = np.random.rand(20)  

plt.scatter(x, y, color='none', edgecolors='black')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Scatter Plot with Empty Circles")

plt.show()
```

**3. Write a Python program to draw a scatter plot using random distributions to generate balls of
different sizes.**

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 500

plt.scatter(x, y, s=sizes, color='blue', alpha=0.5, edgecolors='black')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Scatter Plot with Varying Ball Sizes")

plt.show()
```

**5. Write a Python program to draw a scatter plot for three different groups camparing weights
and heights.**

```python
weights1 = np.random.randint(50, 80, 10)
heights1 = np.random.randint(150, 180, 10)
weights2 = np.random.randint(60, 90, 10)
heights2 = np.random.randint(160, 190, 10)
weights3 = np.random.randint(70, 100, 10)
heights3 = np.random.randint(170, 200, 10)

plt.scatter(weights1, heights1, color='red', label='Group 1')
plt.scatter(weights2, heights2, color='blue', label='Group 2')
plt.scatter(weights3, heights3, color='green', label='Group 3')

plt.xlabel("Weight (kg)")
plt.ylabel("Height (cm)")
plt.title("Scatter Plot of Weights vs Heights for Three Groups")
plt.legend()
plt.show()
```
