Languages. Attach a text label above each bar displaying its popularity (float value).<br>**

**7. Write a Python programming to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.<br>**
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>

```python
import matplotlib.pyplot as plt
import numpy as np  

languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]


x_pos = np.array([1, 3, 5, 7, 9, 11])

plt.bar(x_pos, popularity, color='skyblue', edgecolor='black', width=1.5)

plt.xticks(x_pos, languages)

plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Programming Languages Popularity")

plt.show()

```

**9. Write a Python programming to display a bar chart of the popularity of programming
Languages. Increase bottom margin.<br>**
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>

```python
languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

x_pos = np.arange(len(languages))

plt.bar(x_pos, popularity, color='lightseagreen', edgecolor='black')

plt.xticks(x_pos, languages)

plt.subplots_adjust(bottom=0.3)

plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Programming Languages Popularity")

plt.show()
```

**10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.<br>**
Sample Data:<br>
Means (men) = (22, 30, 35, 35, 26)<br>
Means (women) = (25, 32, 30, 35, 29)<br>

```python
groups = ["A", "B", "C", "D", "E"]
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]

x = np.arange(len(groups))
bar_width = 0.3 

plt.bar(x - bar_width/2, means_men, width=bar_width, color='blue', label="Men")   
plt.bar(x + bar_width/2, means_women, width=bar_width, color='pink', label="Women")  

plt.xticks(x, groups)

plt.xlabel("Groups")
plt.ylabel("Scores")
plt.title("Scores by Group and Gender")

plt.legend()

plt.show()
```

**11. Write a Python program to create bar plot from a DataFrame.
Sample Data Frame:<br>**
a b c d e<br>
2 4,8,5,7,6<br>
4 2,3,4,2,6<br>
6 4,7,4,7,8<br>
8 2,6,4,8,6<br>
10 2,4,3,3,2<br>

```python
import pandas as pd

data = {
    "a": [2, 4, 6, 8, 10],
    "b": [4, 2, 4, 2, 6],
    "c": [8, 3, 7, 6, 4],
    "d": [5, 4, 4, 4, 3],
    "e": [7, 2, 7, 8, 3],
    "f": [6, 6, 8, 6, 2]
}

df = pd.DataFrame(data)

df.plot(x="a", kind="bar", figsize=(8, 5), colormap="viridis")

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Bar Plot from DataFrame")

plt.show()
```

**12. Write a Python program to create bar plots with error bars on the same figure. Sample Date<br>**
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824<br>
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645<br>

```python
categories = ["A", "B", "C", "D"]  
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824] 
std_dev = [0.3314, 0.2278, 0.2836, 0.2645]

plt.bar(categories, mean_velocity, yerr=std_dev, capsize=5, color="skyblue", edgecolor="black", alpha=0.7)

plt.xlabel("Categories")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars")

plt.show()
```

**13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).<br>**
Sample Data<br>
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824<br>
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645<br>

```python
categories = ["A", "B", "C", "D"]
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824] 
std_dev = [0.3314, 0.2278, 0.2836, 0.2645] 

bars = plt.bar(categories, mean_velocity, yerr=std_dev, capsize=5, color="skyblue", edgecolor="black", alpha=0.7)

for bar, value in zip(bars, mean_velocity):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02, str(int(value * 100)), 
             ha='center', fontsize=12, fontweight='bold', color='black')

plt.xlabel("Categories")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars and Labels")

plt.show()
```
