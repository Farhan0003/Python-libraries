**3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1**

```python
with open("test.txt", "w") as file:
    file.write("1 2\n2 4\n3 1")

print("test.txt file created successfully!")
```

```python
x_values = []
y_values = []

with open("test.txt","r")as file:
    for line in file:
        x,y=map(int,line.split())
        x_values.append(x)
        y_values.append(y)

plt.title("Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x_values,y_values)
plt.show()
```

**11. Write a Python program to plot several lines with different format styles in one command
using arrays.**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 1, 4, 9, 16, 25])  
y2 = np.array([0, 1, 2, 3, 4, 5])    
y3 = np.array([0, 1, 8, 27, 64, 125]) 

plt.plot(x, y1, 'r--', label="Quadratic (y = x^2)") 
plt.plot(x, y2, 'g-o', label="Linear (y = x)")       
plt.plot(x, y3, 'b:', label="Cubic (y = x^3)")       

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Multiple Lines with Different Styles")

plt.legend()

plt.show()
```

**12. Write a Python program to create multiple types of charts**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y, marker='o', color='blue')
plt.title("Line Chart")

plt.subplot(2, 2, 2)
plt.bar(x, y, color='green')
plt.title("Bar Chart")

plt.subplot(2, 2, 3)
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")

plt.subplot(2, 2, 4)
labels = ["A", "B", "C", "D", "E"]
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")

plt.tight_layout()
plt.show()
```

**14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.**

```python
dates = ["03-10-16", "04-10-16", "05-10-16", "06-10-16", "07-10-16"]
closing_prices = [772.56, 776.43, 776.47, 776.86, 775.08]

plt.plot(dates, closing_prices, marker='o', linestyle='-', color='green', label="Closing Price")

plt.grid(which='major', linestyle='-', linewidth=0.8, color='blue')

plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=0.5, color='gray')

plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False)

plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.title("Alphabet Inc. Closing Prices (Oct 3-7, 2016)")

plt.legend()

plt.show()
```

**15. Write a Python program to create multiple plots.**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # x^2
y2 = [1, 8, 27, 64, 125]  # x^3

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o', color='blue')
plt.title("Square Function")

plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='s', color='red')
plt.title("Cube Function")

plt.subplot(2, 2, 3)
plt.bar(x, y1, color='green')
plt.title("Bar Chart")

plt.subplot(2, 2, 4)
plt.scatter(x, y2, color='purple')
plt.title("Scatter Plot")

plt.tight_layout()

plt.show()
```
